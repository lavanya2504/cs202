```python
import re
import matplotlib.pyplot as plt

# Load the coverage report
file_path = "C:/Users/Lavanya/Documents/STT/Submission/Lab5/coverage_report_A.txt"
with open(file_path, "r") as file:
    report_content = file.readlines()

# Extract coverage percentages using regex
coverage_pattern = re.compile(r'\s+([\d.]+)%')
coverage_values = []

for line in report_content:
    match = coverage_pattern.search(line)
    if match:
        coverage_values.append(float(match.group(1)))

# Categorize coverage values
low = sum(1 for x in coverage_values if x < 50)
moderate = sum(1 for x in coverage_values if 50 <= x <= 80)
high = sum(1 for x in coverage_values if x > 80)

# Create labels and values for the pie chart
labels = ['Low Coverage (<50%)', 'Moderate Coverage (50-80%)', 'High Coverage (>80%)']
values = [low, moderate, high]

# Generate the pie chart
plt.figure(figsize=(6, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', colors=["#FFB3BA", "#FFDFBA", "#Baffc9"], startangle=140)
plt.title("Coverage Percentage Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Coverage data (example based on your description)
modules = [
    "valid_sudoku",
    "meeting_rooms",
    "longest_common_subsequence",
    "linked_list",
    "insertion_sort"
]

before_coverage = [0, 0, 0, 0, 4]    # coverage before Pynguin
after_coverage  = [14, 17, 8, 44, 7]  # coverage after Pynguin

# Create a grouped bar chart
x = np.arange(len(modules))  # the label locations
width = 0.35                 # the width of the bars

fig, ax = plt.subplots(figsize=(8, 5))
bars_before = ax.bar(x - width/2, before_coverage, width, label='Before Pynguin', color='lightcoral')
bars_after = ax.bar(x + width/2, after_coverage, width, label='After Pynguin', color='skyblue')

# Labels, title, legend
ax.set_ylabel('Coverage (%)')
ax.set_title('Coverage Improvement Using Pynguin')
ax.set_xticks(x, modules, rotation=45, ha='right')
ax.legend()

# Annotate each bar with its coverage value
for bar in bars_before + bars_after:
    height = bar.get_height()
    ax.annotate(
        f'{height}%',
        xy=(bar.get_x() + bar.get_width() / 2, height),
        xytext=(0, 3),  # offset
        textcoords="offset points",
        ha='center', va='bottom',
        fontsize=8
    )

plt.tight_layout()
plt.show()


```
